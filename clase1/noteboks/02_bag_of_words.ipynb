```python
from sklearn.feature_extraction.text import CountVectorizer

# Definir documentos de ejemplo
documents = [
    "El procesamiento del lenguaje natural es fascinante",
    "El lenguaje natural permite la comunicación entre humanos y máquinas",
    "Los modelos de PLN pueden aprender patrones en texto"
]

# Inicializar el vectorizador BoW
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Obtener las palabras únicas (características)
words = vectorizer.get_feature_names_out()

# Convertir la matriz a un array
data = X.toarray()

# Mostrar los resultados
import pandas as pd
pd.DataFrame(data, columns=words)
```
