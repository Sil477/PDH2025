```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Definir un conjunto de documentos de ejemplo
documentos = [
    "El procesamiento del lenguaje natural es una rama de la inteligencia artificial.",
    "Los modelos de lenguaje como Bag of Words son utilizados en NLP.",
    "La tokenización es un paso clave en el procesamiento del lenguaje natural."
]

# Inicializar el vectorizador de bolsa de palabras
vectorizador = CountVectorizer()

# Transformar los documentos en una matriz de características
matriz_bow = vectorizador.fit_transform(documentos)

# Obtener las palabras del vocabulario
vocabulario = vectorizador.get_feature_names_out()

# Convertir la matriz en un DataFrame para visualización
df_bow = pd.DataFrame(matriz_bow.toarray(), columns=vocabulario)

# Mostrar la matriz resultante
df_bow
```
